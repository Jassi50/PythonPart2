#### Design data definitions

I chose to represent the border, date, measure and value. All 4 of these are required. The date will filter down data to 2020, measure will filter data to passenger vehicles. The border and value are directly going to be in the graph.

```python
from cs103 import *
from typing import NamedTuple, List
from enum import Enum
import csv
from matplotlib import pyplot

##################
# Data Definitions

BorderType = Enum('BorderType', ['Mexico', 'Canada'])
# interp. the type of a border as either US-Mexico (Mexico) or US-Canada (Canada)
# examples are redundant for enumerations

# template based on one of (2 cases) and atomic distinct (2x)
@typecheck
def fn_for_border_type(bt: BorderType) -> ...:
    if bt == BorderType.Mexico:
        return ...
    elif bt == BorderType.Canada:
        return ...
    
CrossingType = Enum('CrossingType', ['Trucks', 'TruckContLoaded', 'TruckContEmpty', 'PersVehicles', 'PersVehiclePassengers', 'Buses', 'BusPassengers', 
                                  'Trains','RailContEmpty', 'RailContLoaded', 'TrainPassengers', 'Pedestrians',])
# interp. the crossing method as either Trucks, Truck Containers Loaded (TruckContLoaded), Truck Containers Empty (TruckContEmpty), Personal Vehicles 
# (PersVehicles), Personal Vehicle Passengers (PersVehiclePassengers), Buses, Bus passengers (BusPassengers), Trains, Rail Containers Empty (RailContEmpty), 
# Rail Containers Loaded (RailContLoaded), Train Passengers (TrainPassengers), or Pedestrians.
# examples are redundant for enumerations

# template based on one of (12 cases) and atomic distinct (12x)
@typecheck
def fn_for_crossing_type(ct: CrossingType) -> ...:
    if ct == CrossingType.Trucks:
        return ...
    elif ct == CrossingType.TruckContLoaded:
        return ...    
    elif ct == CrossingType.TruckContEmpty:
        return ...    
    elif ct == CrossingType.PersVehicles:
        return ...    
    elif ct == CrossingType.PersVehiclePassengers:
        return ... 
    elif ct == CrossingType.Buses:
        return ...    
    elif ct == CrossingType.BusPassengers:
        return ...    
    elif ct == CrossingType.Trains:
        return ...    
    elif ct == CrossingType.RailContEmpty:
        return ... 
    elif ct == CrossingType.RailContLoaded:
        return ...    
    elif ct == CrossingType.TrainPassengers:
        return ...    
    elif ct == CrossingType.Pedestrians:
        return ...        

BorderCrossing = NamedTuple('BorderCrossing', [('border', BorderType),
                                   ('date', int),
                                   ('crossing', CrossingType),
                                   ('value', int)])
                                   
# interp. information about a border crossing, with the type of border, date, crossing method (crossing), and a count of how many for that crossing type (value)

C1 = BorderCrossing(BorderType.Mexico, 2023, CrossingType.Trucks, 5823)
C2 = BorderCrossing(BorderType.Mexico, 2009, CrossingType.Trains, 0)
C3 = BorderCrossing(BorderType.Canada, 2023, CrossingType.Pedestrians, 1)
C4 = BorderCrossing(BorderType.Mexico, 2023, CrossingType.Pedestrians, 6383)
C5 = BorderCrossing(BorderType.Canada, 2023, CrossingType.PersVehicles, 294)
C6 = BorderCrossing(BorderType.Mexico, 2023, CrossingType.Pedestrians, 39683)
C7 = BorderCrossing(BorderType.Canada, 2020, CrossingType.PersVehicles, 2000)
C8 = BorderCrossing(BorderType.Mexico, 2020, CrossingType.PersVehicles, 3000)
C9 = BorderCrossing(BorderType.Canada, 2020, CrossingType.PersVehicles, 570)
C10 = BorderCrossing(BorderType.Canada, 2020, CrossingType.PersVehicles, 1022)
C11 = BorderCrossing(BorderType.Canada, 2020, CrossingType.PersVehicles, 491)
C12 = BorderCrossing(BorderType.Mexico, 2020, CrossingType.PersVehicles, 63397)

# template based on compound (4 fields) and reference rule (once)
@typecheck
def fn_for_BorderCrossing(bc: BorderCrossing) -> ...:
    return ...(fn_for_border_type(bc.border),
               bc.date,
               fn_for_crossing_type(bc.crossing),
               bc.value)                     
       
# List[BorderCrossing]
# interp. a list of Border Crossing data
LOC0 = []
LOC1 = [C1, C2, C3]
LOC2 = [C4, C5, C6] 
LOC3 = [C7, C5, C3, C2, C1]
LOC4 = [C7, C8, C1]
LOC5 = [C1, C2, C3, C9, C10]
LOC6 = [C4, C5, C6, C11, C12]

# template based on arbitrary-sized and reference rule
@typecheck
def fn_for_lobc(lobc: List[BorderCrossing]) -> ...:
    # description of the acc
    acc = ...      # type: ...
    for bc in lobc:
        acc = ...(fn_for_BorderCrossing(bc), acc)
    
    return ...(acc)
```

### Step 2b and 2c: Building
#### Design a function to read the information and store it as data in your program
#### Design functions to analyze the data

```python
###########
# Functions

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename and plot a pie chart showing the proportion of personal vehicles that crossed each border(Canada and Mexico) into the US in 2020 
    """
    # Template from HtDAP, based on function composition 
    return plot_personal_vehicles_vs_border_type(read(filename)) 
    
@typecheck
def read(filename: str) -> List[BorderCrossing]:
    """    
    reads information from the specified file and returns a list of BorderCrossings
    """
    #return []  #stub
    # Template from HtDAP
    # loc contains the result so far
    loc = [] # type: List[Consumed]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            bc = BorderCrossing(parse_border_type(row[3]),      # border
                         parse_date(row[4]),                    # date                         
                         parse_crossing_type(row[5]),           # crossing
                         parse_int(row[6]))                     # value
            loc.append(bc)
    
    return loc

@typecheck
def parse_date(year: str) -> int:
    """
    convert year into the appropriate foramt(xxxx).
    """
    # return 2020 # stub
    # template based on Atomic Non-Distinct
    return int(year[0:4])

@typecheck
def parse_border_type(border_type: str) -> BorderType:
    """
    convert border_type (which must be one of "US-Mexico Border", "US-Canada Border") into
    the corresponding BorderType.
    """
    # return BorderType.Mexico  #stub
    # template from atomic non-distinct (with guidance from BorderType)
    if border_type == "US-Mexico Border":
        return BorderType.Mexico
    elif border_type == "US-Canada Border":
        return BorderType.Canada

@typecheck
def parse_crossing_type(crossing_type: str) -> CrossingType:
    """
    convert crossing_type (which must be one of "Bus Passengers", "Buses", "Pedestrians", "Personal Vehicle Passengers", "Personal Vehicles", "Rail Containers Empty",
    "Rail Containers Loaded", "Train Passengers", "Trains", "Truck Containers Empty", "Truck Containers Loaded", "Trucks") into
    the corresponding CrossingType.
    """
    # return CrossingType.BusPassengers  #stub
    # template from atomic non-distinct (with guidance from CrossingType)
    if crossing_type == "Bus Passengers":
        return CrossingType.BusPassengers
    elif crossing_type == "Buses":
        return CrossingType.Buses    
    elif crossing_type == "Pedestrians":
        return CrossingType.Pedestrians 
    elif crossing_type == "Personal Vehicle Passengers":
        return CrossingType.PersVehiclePassengers   
    elif crossing_type == "Personal Vehicles":
        return CrossingType.PersVehicles    
    elif crossing_type == "Rail Containers Empty":
        return CrossingType.RailContEmpty 
    elif crossing_type == "Rail Containers Loaded":
        return CrossingType.RailContLoaded 
    elif crossing_type == "Train Passengers":
        return CrossingType.TrainPassengers    
    elif crossing_type == "Trains":
        return CrossingType.Trains 
    elif crossing_type == "Truck Containers Empty":
        return CrossingType.TruckContEmpty 
    elif crossing_type == "Truck Containers Loaded":
        return CrossingType.TruckContLoaded    
    elif crossing_type == "Trucks":
        return CrossingType.Trucks         
    
@typecheck
def plot_personal_vehicles_vs_border_type(border_crossing_data: List[BorderCrossing]) -> None:
    """
    Plots the total number of personal vehicles that crossed each border type in 2020.  
    
    Draws the plot and returns None.
    """
    #return None  #stub
    # Template based on visualization
    # We divided off three distinct subtasks into helper functions: 
    # + filtering the data down to bc in 2020 
    # + filtering the data down to bc that used personal vehicle crossings
    # + getting the number of crossings for the US-Canada Border
    # + getting the number of crossings for the US-Mexico Border  
    data_filtered_by_year = year_filter(border_crossing_data)
    data_filtered_by_crossing = crossing_filter(data_filtered_by_year, CrossingType.PersVehicles)
    canada_crossing_proportion = canada_crossing_amount(data_filtered_by_crossing, BorderType.Canada)
    mexico_crossing_proportion = mexico_crossing_amount(data_filtered_by_crossing, BorderType.Mexico)

    pie_chart_labels = ["Canada", "Mexico"]
    pie_chart_fractions = [canada_crossing_proportion, mexico_crossing_proportion]
    pie_chart_explode = [0, 0]
    
    if pie_chart_fractions == [0, 0]: 
        return None
        
    return show_pie_chart(pie_chart_labels, pie_chart_fractions, pie_chart_explode)
    
#  HELPER FUNCTION FOR STEP 1

@typecheck
def year_filter(lobc: List[BorderCrossing]) -> List[BorderCrossing]:
    """
    takes lobc and returns a list of bc that occurred in 2020
    """
    # return [] # stub
    # template from List[BorderCrossing] with 1 additional parameter (y: 2020) 
    # the border crossings that occured in 2020 so far
    acc = [] # type: List[BorderCrossing]
    for bc in lobc:
        if is_year(bc):
            acc.append(bc)
    return (acc)

@typecheck
def is_year(bc: BorderCrossing) -> bool:
    """
    returns True if bc.date == 2020, False otherwise
    """
    # return True # stub
    # template based on BorderCrossing
    return bc.date == 2020

#  HELPER FUNCTION FOR STEP 2

@typecheck
def crossing_filter(lobc: List[BorderCrossing], ct: CrossingType) -> List[BorderCrossing]:
    """
    takes lobc and returns a list of bc that that used personal vehicles to cross
    """
    # return [] # stub
    # template from List[BorderCrossing] with 1 additional parameter (ct: CrossingType) 
    # the border crossings that occured that used personal vehicles to cross so far
    acc = [] # type: List[BorderCrossing]
    for bc in lobc:
        if is_crossing_type(bc, ct):
            acc.append(bc)
    return (acc)

@typecheck
def is_crossing_type(bc: BorderCrossing , ct: CrossingType) -> bool:
    """
    returns True if bc.crossing == PersVehicles, False otherwise
    """
    # return True # stub
    # template based on BorderCrossing
    return bc.crossing == ct.PersVehicles

#  HELPER FUNCTION FOR STEP 3

@typecheck
def canada_crossing_amount(lobc: List[BorderCrossing], bt: BorderType) -> int:
    """
    takes lobc and returns the number of border crossings that were from Canada
    """
    # return 0 # stub
    # template from List[BorderCrossing] with 1 additional parameter (bt: BorderType) 
    # the number of border crossings that were from Canada so far
    acc = 0 # type: int
    for bc in lobc:
        if is_canada(bc, bt):
            acc = acc + bc.value
    return (acc)

@typecheck
def is_canada(bc: BorderCrossing, bt: BorderType) -> bool:
    """ 
    returns True if bc.border == Canada
    """
    # return True # stub
    # template based on BorderCrossing
    return bc.border== bt.Canada

#  HELPER FUNCTION FOR STEP 4

@typecheck
def mexico_crossing_amount(lobc: List[BorderCrossing], bt: BorderType) -> int:
    """
    takes lobc and returns the number of border crossings that were from Mexico
    """
    # return 0 # stub
    # template from List[BorderCrossing] with 1 additional parameter (bt: BorderCrossing) 
    # the number of border crossings that were from Mexico so far
    acc = 0 # type: int
    for bc in lobc:
        if is_mexico(bc, bt):
            acc = acc + bc.value
    return (acc)

@typecheck
def is_mexico(bc: BorderCrossing, bt: BorderType) -> bool:
    """ 
    returns True if bc.border == Mecico
    """
    # return True # stub
    # template based on BorderCrossing
    return bc.border == bt.Mexico

@typecheck
def show_pie_chart(labels: List[str], fractions: List[int], explode: List[float]) -> None:
    """
    show a pie chart with the given labels and corresponding fractions.
    
    "explode" (set out from the circle) each piece of the pie as specified in explode.
    
    fractions must each be in the range [0,100] and must sum to 100
    (and so must have at least one element).
    
    labels and fractions must have the same length.
    
    returns None
    """
    # return None  #stub
    # Template based on visualization

    # specify the percentages, then whether to explode each piece, then the labels, then the formatting
    # for the percentages on the chart, and finally whether you want a shadow or not. If you
    # remove autopct, the percentage will not show on each piece of the pie
    pyplot.pie(fractions, explode=explode, labels=labels, autopct='%.0f%%', shadow=False)
    
    pyplot.title("Inbound US Personal Vehicle Border Crossings by Country in 2020")
    
    pyplot.show()
    
    return None  


start_testing()

expect(main('Border_Crossing_Entry_Data.csv'), None)  
expect(main('test1.csv'), None)  
expect(main('test2.csv'), None)  

expect(read("test1.csv"), LOC5)
expect(read("test2.csv"), LOC6)

expect(parse_date("2023-02-01"), 2023)
expect(parse_date("2009-08-01"), 2009)

expect(parse_border_type("US-Mexico Border"), BorderType.Mexico)
expect(parse_border_type("US-Canada Border"), BorderType.Canada)

expect(parse_crossing_type("Bus Passengers"), CrossingType.BusPassengers)
expect(parse_crossing_type("Buses"), CrossingType.Buses)
expect(parse_crossing_type("Pedestrians"), CrossingType.Pedestrians)
expect(parse_crossing_type("Personal Vehicle Passengers"), CrossingType.PersVehiclePassengers)
expect(parse_crossing_type("Personal Vehicles"), CrossingType.PersVehicles)
expect(parse_crossing_type("Rail Containers Empty"), CrossingType.RailContEmpty)
expect(parse_crossing_type("Rail Containers Loaded"), CrossingType.RailContLoaded)
expect(parse_crossing_type("Train Passengers"), CrossingType.TrainPassengers)
expect(parse_crossing_type("Trains"), CrossingType.Trains)
expect(parse_crossing_type("Truck Containers Empty"), CrossingType.TruckContEmpty)
expect(parse_crossing_type("Truck Containers Loaded"), CrossingType.TruckContLoaded)
expect(parse_crossing_type("Trucks"), CrossingType.Trucks)

expect(plot_personal_vehicles_vs_border_type(LOC4), None)
expect(plot_personal_vehicles_vs_border_type(LOC1), None)

expect(year_filter(LOC1), [])
expect(year_filter(LOC3), [C7])

expect(is_year(C1), False)
expect(is_year(C7), True)

expect(crossing_filter(LOC1, CrossingType.PersVehicles), [])
expect(crossing_filter(LOC3, CrossingType.PersVehicles), [C7, C5])

expect(is_crossing_type(C1, CrossingType.PersVehicles), False) 
expect(is_crossing_type(C7, CrossingType.PersVehicles), True) 

expect(canada_crossing_amount(LOC1, BorderType.Canada), 1)
expect(canada_crossing_amount(LOC3, BorderType.Canada), 2295)

expect(is_canada(C1, BorderType.Canada), False)
expect(is_canada(C7, BorderType.Canada), True)

expect(mexico_crossing_amount(LOC1, BorderType.Mexico), 5823)
expect(mexico_crossing_amount(LOC2, BorderType.Mexico), 46066)

expect(is_mexico(C1, BorderType.Mexico), True)
expect(is_mexico(C7, BorderType.Mexico), False)

expect(show_pie_chart(['Canada'], [100], [0]), None)
expect(show_pie_chart(['Canada', 'Mexico'], [50, 50], [0, 0]), None)

summary()
```
