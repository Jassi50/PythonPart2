## Problem 1

Design a function that takes a departing station and list of trips and returns the shortest trip duration for users that are subscribers. 

In particular, the **function MUST operate in three steps** First, it will find the trips from the given departure station, then it will find which of those trips were taken by users that are subscribers and lastly it will find shortest trip duration.

For example given `LOT1`  and station `Station.wellington`, the function should return `152`

Use the following hints to guide the design of you functions to find shortest trip duration:

   1. Your `List[Trip]` can be empty either as a result of the filtering process or you were given a list of customers only (not subscribers) 
   2. When thinking about the output of your function, think about the data type allows you to represent missing data or special values.
    Given an empty list of trips `[]`, why would returning `0`  be misleading? 

Remember to follow the HtDF recipe and the helper rules.

```python
# Your solution goes here
@typecheck
def shortest_trip_duration_for_subscribers(s: Station, lot: List[Trip]) -> Optional[int]:
    """
    takes a departing station (s) and list of trips (lot) and returns the shortest trip duration in seconds (int) for users that are subscribers
    """
    # return 0 # stub
    # template based on composition
    # PLAN 
    # 1. Filter lot by Station where departing station is (s)
    # 2. Filter lot by which trips are taken by users that are subscribers
    # 3. Filter lot by which trip is shortest and return its duration 
    return shortest_trip_length(users_subscribers(departing_station(s, lot)))
    
#  HELPER FUNCTION FOR STEP 1

@typecheck 
def departing_station(s: Station, lot: List[Trip]) -> List[Trip]:
    """
    takes lot and s, and returns a list of trips that departed from s
    """
    # return [] # stub
    # template based on List[Trip]
    # the trips that departed from s so far
    acc = [] # type: List[Trip]
    for t in lot:
        if is_station(t, s):
            acc.append(t)
    return (acc)

@typecheck
def is_station(t: Trip, s: Station) -> bool:
    """
    returns True if t.from_station == s, False otherwise
    """
    # return True # stub
    # template based on Trip
    return t.from_station == s

#  HELPER FUNCTION FOR STEP 2

@typecheck
def users_subscribers(lot: List[Trip]) -> List[Trip]:
    """
    takes lot and returns a list of trips where the users are subscribers
    """
    # return [] # stub
    # template based on List[Trip]
    # the trips where the users are subscribers so far
    acc = [] # type: List[Trip]
    for t in lot:
        if is_user(t):
            acc.append(t)
    return (acc) 

@typecheck
def is_user(t: Trip) -> bool:
    """
    returns True if t.user_type == True, False otherwise
    """
    # return True # stub
    # template based on Trip
    return t.user_type == True
 
#  HELPER FUNCTION FOR STEP 3

@typecheck
def shortest_trip_length(lot: List[Trip]) -> Optional[int]:
    """
    takes lot, returns the length in seconds of the shortest trip in the list
    """
    # return 0 # stub
    # template based on composition
    # PLAN
    # 1. Filter lot and find the shortest trip 
    # 2. Determine the length of the shortest trip
    if(lot == []):
        return None
    else: 
        return length(shortest_trip(lot))

#  HELPER FUNCTION FOR STEP 3.1

@typecheck
def shortest_trip(lot: List[Trip]) -> Trip:
    """
    takes lot, returns the shortest trip in the list
    """
    # return T1 # stub
    # the shortest trip so far, start with the first trip in the list
    acc = lot[0] # type: Trip
    for t in lot:
        if is_shortest_trip(t, acc):
            acc = t    
    return (acc)

@typecheck
def is_shortest_trip(t1: Trip, t2: Trip) -> bool:
    """
    returns True if t1 is shorter than t2, False otherwise
    """
    # return True # stub
    # template based on Trip
    return t1.duration < t2.duration

#  HELPER FUNCTION FOR STEP 3.2

@typecheck
def length(t:Trip) -> int:
    """
    Takes a trip and returns the duration of the trip in seconds
    """
    # return 0 # stub
    # template based on Trip
    return t.duration

# Begin testing
start_testing()

expect(shortest_trip_duration_for_subscribers(Station.wellington, LOT0), None)
expect(shortest_trip_duration_for_subscribers(Station.wellington, LOT1), 152)
expect(shortest_trip_duration_for_subscribers(Station.ashland, LOT2), 226)
expect(shortest_trip_duration_for_subscribers(Station.california, LOT3), 752)

expect(departing_station(Station.wellington, LOT1), [T1, T2, T10])
expect(departing_station(Station.ashland, LOT2), [T3, T4])
expect(departing_station(Station.california, LOT3), [T5, T6])

expect(is_station(T1, Station.wellington), True)
expect(is_station(T2, Station.ashland), False)
expect(is_station(T3, Station.california), False)

expect(users_subscribers(LOT1), [T1, T3, T6, T7, T8, T10])
expect(users_subscribers(LOT2), [T1, T3])
expect(users_subscribers(LOT3), [T6, T7, T8])

expect(is_user(T1), True)
expect(is_user(T2), False)
expect(is_user(T3), True)

expect(shortest_trip_length(LOT1), 152)
expect(shortest_trip_length(LOT2), 226)
expect(shortest_trip_length(LOT3), 153)

expect(shortest_trip(LOT1), T10)
expect(shortest_trip(LOT2), T3)
expect(shortest_trip(LOT3), T8)

expect(is_shortest_trip(T1, T2), False)
expect(is_shortest_trip(T3, T4), True)
expect(is_shortest_trip(T5, T6), True)

expect(length(T1), 926)
expect(length(T2), 497)
expect(length(T3), 226)

# Finish testing
summary()
```
